```python
from PIL import Image
import torch
from transformers import AutoProcessor, Qwen2_5_VLForConditionalGeneration
from pydantic import BaseModel, Field
from typing import List, Dict, Any
from qwen_vl_utils import process_vision_info
import json
from TTS.api import TTS
import gradio as gr

# Define Pydantic models to structure the input and output data
class EmailExtractRequest(BaseModel):
    image_path: str  # Path to the image containing the email
    role: str = Field(default="user")
    message_content: str = Field(
        default="You are given an image that shows an email. "
                "Extract the following fields in JSON format: sender, subject, and body. "
                "Ensure your JSON output is valid."
    )

class EmailField(BaseModel):
    sender: str
    subject: str
    body: str

class EmailExtractResponse(BaseModel):
    email: EmailField

# Define constants for model and device configuration
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"

# Load model and processor
model = Qwen2_5_VLForConditionalGeneration.from_pretrained(
    "Qwen/Qwen2.5-VL-3B-Instruct",
    torch_dtype=torch.bfloat16,
    attn_implementation="flash_attention_2",
    device_map="auto",
)

processor = AutoProcessor.from_pretrained("Qwen/Qwen2.5-VL-3B-Instruct", use_fast=True)

# Function to extract email details from the image
def extract_email_details(image_path: str) -> dict:
    # Create an instance of the request model
    email_request = EmailExtractRequest(image_path=image_path)

    # Define messages for model input
    messages = [
        {
            "role": email_request.role,
            "content": [
                {
                    "type": "image",
                    "image": email_request.image_path,
                },
                {
                    "type": "text",
                    "text": email_request.message_content,
                },
            ],
        }
    ]

    # Prepare the input text using processor
    text = processor.apply_chat_template(
        messages, tokenize=False, add_generation_prompt=True
    )

    # Process vision data
    image_inputs, video_inputs = process_vision_info(messages)

    # Prepare inputs for the model
    inputs = processor(
        text=[text],
        images=image_inputs,
        videos=video_inputs,
        padding=True,
        return_tensors="pt",
    )
    inputs = inputs.to(DEVICE)

    # Generate model output
    generated_ids = model.generate(**inputs, max_new_tokens=1000)
    generated_ids_trimmed = [
        out_ids[len(in_ids):] for in_ids, out_ids in zip(inputs.input_ids, generated_ids)
    ]

    output_text = processor.batch_decode(
        generated_ids_trimmed, skip_special_tokens=True, clean_up_tokenization_spaces=False
    )

    # Parse output into structured format
    response_data = parse_json_output(output_text[0])
    return response_data

# Function to parse JSON output from the model's response
def parse_json_output(output_text: str) -> Dict[str, Any]:
    try:
        # Parse the model's output (assuming it's in JSON format)
        from json_repair import repair_json
        parsed_output = repair_json(output_text)
        # Ensure the parsed output contains 'sender', 'subject', and 'body'
        if not all(field in parsed_output for field in ['sender', 'subject', 'body']):
            raise ValueError(f"Invalid JSON structure: {parsed_output}")
        return parsed_output
    except Exception as e:
        raise ValueError(f"Failed to parse output JSON: {str(e)}")

# Function to summarize the extracted email
def summarize_email(email_data) -> str:
    summary_prompt = (
        "Summarize the following email concisely while maintaining key information:\n"
        f"Sender: {email_data['sender']}\n"
        f"Subject: {email_data['subject']}\n"
        f"Body: {email_data['body']}\n"
    )

    messages = [{"role": "user", "content": [{"type": "text", "text": summary_prompt}]}]
    text = processor.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
    inputs = processor(text=[text], padding=True, return_tensors="pt").to(DEVICE)

    generated_ids = model.generate(**inputs, max_new_tokens=200)
    generated_text = processor.batch_decode(generated_ids, skip_special_tokens=True, clean_up_tokenization_spaces=False)

    return generated_text[0].split("assistant\n", 1)[-1].strip()

# Function to convert text to audio
def text_to_audio(text, output_file="output.wav"):
    """
    Converts text to speech using Coqui TTS and saves it as an audio file.
    """
    try:
        # Initialize the TTS model
        device = "cuda" if torch.cuda.is_available() else "cpu"
        tts = TTS(model_name="tts_models/en/ljspeech/tacotron2-DDC", progress_bar=False, gpu=False).to(device)

        # Generate speech and save to file
        tts.tts_to_file(text=text, file_path=output_file)
        print(f"Audio saved to {output_file}")
    except Exception as e:
        print(f"Error converting text to audio: {e}")

# Gradio app function
def process_email_screenshot(image):
    # Step 1: Extract email details
    email_data = extract_email_details(image)
    email_data = json.loads(email_data)

    # Step 2: Summarize the email
    email_summary = summarize_email(email_data)
    # email_summary = f"Sender: {email_data['sender']}. Subject: {email_data['subject']}. Summary:{email_summary}"
    email_summary_write_up = f"Sender: {email_data['sender']}\nSubject: {email_data['subject']}\nSummary: {email_summary}"

    # Step 3: Add sender and subject to the audio summary
    audio_summary_text = f"Sender: {email_data['sender']}. Subject: {email_data['subject']}. Summary: {email_summary}"

    # Step 4: Convert summary to audio
    text_to_audio(audio_summary_text, output_file="email_summary.wav")

    return email_summary_write_up, "email_summary.wav"

# Gradio interface
with gr.Blocks() as demo:
    gr.Markdown("# AI-Powered Email Reader")
    gr.Markdown("Upload a screenshot of an email to get a written and audio summary.")

    with gr.Row():
        image_input = gr.Image(label="Upload Email Screenshot", type="filepath")
        text_output = gr.Textbox(label="Email Summary")
        audio_output = gr.Audio(label="Audio Summary")

    submit_button = gr.Button("Process Email")
    submit_button.click(
        fn=process_email_screenshot,
        inputs=image_input,
        outputs=[text_output, audio_output]
    )

# Launch the Gradio app
demo.launch(debug = True)
```
